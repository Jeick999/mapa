```python
# Visualizador de Unidades Educativas en TulcÃ¡n - Colores Modificados

# InstalaciÃ³n de paquetes necesarios
!pip install geopandas folium matplotlib -q

# ImportaciÃ³n de librerÃ­as
import geopandas as gpd
import folium
import json
import requests
import matplotlib.pyplot as plt
import tempfile
import os

# URL del archivo GeoJSON
url = "https://raw.githubusercontent.com/Jeick999/mapa/refs/heads/main/Jeick69-map%20(1).geojson"

# Descargar y cargar los datos
response = requests.get(url)
geojson_data = json.loads(response.text)

# Convertir a GeoDataFrame con manejo de propiedades
def extract_properties(feature):
    """Extrae propiedades de manera segura."""
    properties = feature.get('properties', {})
    title = properties.get('title', 'Sin tÃ­tulo')
    fill_color = properties.get('fill', '#3cc954')
    marker_color = properties.get('marker-color', '#3cc954')
    return {
        'title': title,
        'fill': fill_color,
        'marker-color': marker_color
    }

# Modificar features para incluir propiedades extraÃ­das
for feature in geojson_data['features']:
    feature['properties'] = extract_properties(feature)

# Convertir a GeoDataFrame
gdf = gpd.GeoDataFrame.from_features(geojson_data['features'])
gdf.crs = "EPSG:4326"

# Separar los datos por tipo de geometrÃ­a
points_gdf = gdf[gdf.geometry.type == 'Point']
polygons_gdf = gdf[gdf.geometry.type == 'Polygon']

# Identificar el lÃ­mite de la ciudad
city_boundary = polygons_gdf[polygons_gdf['title'] == 'New item']
schools_polygons = polygons_gdf[polygons_gdf['title'] != 'New item']

# Crear mapa interactivo con Folium - TEMA CLARO
m = folium.Map(
    location=[0.805, -77.72],
    zoom_start=14,
    control_scale=True,
    tiles='CartoDB positron'  # Cambiado a tema claro
)

# AÃ±adir capa de satÃ©lite personalizada
folium.TileLayer(
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='Esri',
    name='SatÃ©lite',
    overlay=False
).add_to(m)

# AÃ±adir capa de mapa colorido y brillante
folium.TileLayer(
    tiles='https://tiles.stadiamaps.com/tiles/alidade_smooth/{z}/{x}/{y}.png',
    attr='Stadia Maps',
    name='Mapa Claro Suave',
    overlay=False
).add_to(m)

# Crear grupos de capas
city_layer = folium.FeatureGroup(name="LÃ­mite de la Ciudad")
schools_polygons_layer = folium.FeatureGroup(name="Ãreas de Instituciones")
schools_points_layer = folium.FeatureGroup(name="UbicaciÃ³n de Instituciones")

# AÃ±adir lÃ­mite de la ciudad - COLORES BRILLANTES Y CLAROS
for idx, row in city_boundary.iterrows():
    sim_geo = gpd.GeoSeries(row['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                        style_function=lambda x: {
                            'fillColor': '#FFD700',    # Dorado brillante
                            'color': '#FF8C00',        # Naranja dorado
                            'weight': 4,
                            'fillOpacity': 0.6,
                            'dashArray': '10, 5'
                        })
    folium.Tooltip("Ciudad de TulcÃ¡n").add_to(geo_j)
    geo_j.add_to(city_layer)

# Paleta de colores brillantes y claros para polÃ­gonos de instituciones
polygon_colors = ['#FF1493', '#00CED1', '#32CD32', '#FF4500', '#9932CC', '#00FA9A', '#FF69B4', '#1E90FF']

# AÃ±adir Centros de Salud y laboratorios clinicos  - COLORES MODIFICADOS
for idx, row in schools_polygons.iterrows():
    # Asignar color basado en el Ã­ndice
    fill_color = polygon_colors[idx % len(polygon_colors)]

    sim_geo = gpd.GeoSeries(row['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                          style_function=lambda x, color=fill_color: {
                              'fillColor': color,
                              'color': '#000000',        # Borde negro para contraste
                              'weight': 3,
                              'fillOpacity': 0.9,
                              'dashArray': '1, 1'
                          })
    folium.Tooltip(row['title']).add_to(geo_j)
    folium.Popup(f"<strong>{row['title']}</strong><br>Ãrea del Centro Educativo", max_width=300).add_to(geo_j)
    geo_j.add_to(schools_polygons_layer)

# Colores brillantes para los marcadores de puntos
point_colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen', 'gray', 'black']
point_icons = ['graduation-cap', 'school', 'university', 'book', 'pencil', 'edit', 'home', 'building']

# AÃ±adir puntos de Centros de Salud y laboratorios clinicos  - COLORES Y ICONOS MODIFICADOS
for idx, row in points_gdf.iterrows():
    lat, lon = row.geometry.y, row.geometry.x

    # Asignar color e icono basado en el Ã­ndice
    folium_color = point_colors[idx % len(point_colors)]
    icon_type = point_icons[idx % len(point_icons)]

    marker = folium.Marker(
        location=[lat, lon],
        tooltip=row['title'],
        popup=f"<strong>{row['title']}</strong><br>Centro Educativo",
        icon=folium.Icon(
            color=folium_color,
            icon=icon_type,
            prefix='fa'
        )
    )
    marker.add_to(schools_points_layer)

# AÃ±adir capas al mapa
city_layer.add_to(m)
schools_polygons_layer.add_to(m)
schools_points_layer.add_to(m)

# AÃ±adir control de capas
folium.LayerControl().add_to(m)

# AÃ±adir tÃ­tulo al mapa - ESTILO BRILLANTE
title_html = '''
<h3 align="center" style="font-size:20px; color:#FFFFFF; background: linear-gradient(45deg, #FF6B6B, #4ECDC4, #45B7D1, #96CEB4, #FFEAA7); padding:15px; border-radius:15px; margin:15px; box-shadow: 0 4px 8px rgba(0,0,0,0.3); text-shadow: 2px 2px 4px rgba(0,0,0,0.5);"><b>ð UNIDADES EDUCATIVAS EN TULCÃN ð«</b></h3>
'''
m.get_root().html.add_child(folium.Element(title_html))

# AÃ±adir leyenda - COLORES BRILLANTES
legend_html = '''
<div style="position: fixed;
     bottom: 50px; left: 50px; width: 280px; height: auto;
     border:4px solid #FF1493; z-index:9999; font-size:13px;
     background: linear-gradient(135deg, #FFD700 0%, #FF6B6B 50%, #4ECDC4 100%);
     color: #000000; padding: 20px; border-radius: 15px;
     box-shadow: 0 6px 12px rgba(0, 0, 0, 0.3);
     ">
     <p style="margin-top: 0; margin-bottom: 18px; font-size: 16px; font-weight: bold;"><b>ð LEYENDA</b></p>
     <div style="margin-bottom: 12px;">
         <i class="fa fa-graduation-cap fa-2x" style="color:#FF1493;"></i> <strong>UbicaciÃ³n de Instituciones</strong>
     </div>
     <div style="margin-bottom: 12px;">
         <div style="width:30px;height:30px;background-color:#FFD700;border:3px solid #FF8C00;display:inline-block;margin-right:10px;border-radius:5px;"></div>
         <strong>LÃ­mite de la Ciudad</strong>
     </div>
     <div>
         <div style="width:30px;height:30px;background:linear-gradient(45deg, #FF1493, #00CED1);border:3px solid #000000;display:inline-block;margin-right:10px;border-radius:5px;"></div>
         <strong>Ãreas de Instituciones</strong>
     </div>
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# Guardar el mapa interactivo
m.save('mapa_unidades_educativas_tulcan_modificado.html')

# Crear mapa estÃ¡tico - COLORES BRILLANTES Y CLAROS
plt.figure(figsize=(18, 18))
plt.style.use('default')  # Estilo claro
ax = plt.gca()

# Configurar fondo claro
ax.set_facecolor('#F8F9FA')

# Dibujar el lÃ­mite de la ciudad
city_boundary.plot(ax=ax, color='#FFD700', alpha=0.7, edgecolor='#FF8C00', linewidth=4)

# Dibujar polÃ­gonos de Centros de Salud y laboratorios clinicos  con colores brillantes
for idx, row in schools_polygons.iterrows():
    color = polygon_colors[idx % len(polygon_colors)]
    gpd.GeoDataFrame([row]).plot(ax=ax, color=color, alpha=0.9, edgecolor='black', linewidth=3)

# Dibujar puntos de instituciones con colores muy brillantes
point_colors_static = ['#FF0000', '#0000FF', '#00FF00', '#FF00FF', '#FFA500', '#8A2BE2', '#DC143C', '#00CED1']
for idx, row in points_gdf.iterrows():
    color = point_colors_static[idx % len(point_colors_static)]
    gpd.GeoDataFrame([row]).plot(ax=ax, color=color, markersize=150, edgecolor='black', linewidth=3)

# AÃ±adir etiquetas a los puntos con colores brillantes
for idx, row in points_gdf.iterrows():
    plt.annotate(text=row['title'],
                xy=(row.geometry.x, row.geometry.y),
                xytext=(8, 8),
                textcoords="offset points",
                fontsize=12,
                color='black',
                fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.5", facecolor='yellow', alpha=0.9, edgecolor='black', linewidth=2))

plt.title('Centros de Salud y laboratorios clinicos de la ciudad de tulcÃ¡n', fontsize=24, color='##FF1493)', fontweight='bold', pad=30)
plt.axis('off')
plt.tight_layout()
plt.savefig('mapa_estatico_tulcan_brillante.png', dpi=300, facecolor='#F8F9FA', edgecolor='none')

```
